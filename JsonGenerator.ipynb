```python
import pandas as pd
import json
```

```python
def generaJson(ruta,nombre_archivo):
    directorio = ruta
    salida = nombre_archivo+".json"
    entrada = nombre_archivo+".tsv"    
    corpus_train = pd.read_csv(directorio+entrada,delimiter='\t',encoding='utf-8')
    text = corpus_train[corpus_train.columns[1]]
    hs = corpus_train[corpus_train.columns[2]]
    text_hs= list(zip(text,hs))   
    data = {}  
    data[None] = []  
    for (tuit, hs) in text_hs:    
        if hs == 1:
            data[None].append({  
            "text": str(tuit),
            "klass": "HS",
            "value": int(hs)
            })
        else:        
            data[None].append({  
            "text": str(tuit),
            "klass": "NHS",
            "value": int(hs)})
    with open(directorio+salida, 'w') as outfile:
        json.dump(data, outfile)
    return
```

```python
generaJson("corpus/public_development_en_TaskA/","train_en_cleanA")
```

```python
generaJson("corpus/public_development_esTaskA/","train_es_cleanA")
```
